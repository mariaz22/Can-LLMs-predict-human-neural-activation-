## 3. Încarcă Datele Pereira2018

```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

class PereiraDataLoader:
    """
    Încarcă datele Pereira2018 (stimuli + fMRI) în format compatibil cu neural-nlp.
    """

    def __init__(self, data_dir):
        self.data_dir = data_dir
        self.experiments = {
            '243sentences': 'examples_243sentences.mat',
            '384sentences': 'examples_384sentences.mat'
        }

    def load_stimuli(self):
        """
        Încarcă propozițiile din ambele experimente.

        Returns:
            Dict cu experiment -> lista de propoziții
        """
        stimuli = {}

        for exp_name in ['243sentences', '384sentences']:
            filepath = os.path.join(self.data_dir, f'stimuli_{exp_name}.txt')
            if os.path.exists(filepath):
                with open(filepath, 'r') as f:
                    sentences = [line.strip() for line in f if line.strip()]
                stimuli[exp_name] = sentences
                print(f"Loaded {len(sentences)} sentences from {exp_name}")

        return stimuli

    def load_fmri_data(self, max_subjects=None):
        """
        Încarcă datele fMRI din fișierele .mat.

        Args:
            max_subjects: Numărul maxim de subiecți de încărcat (None = toți)

        Returns:
            Dict cu structura:
            {
                'experiment_name': {
                    'sentences': [...],
                    'fmri': np.array [n_sentences, n_voxels_total],
                    'subjects': [...],
                    'voxel_to_subject': [...]
                }
            }
        """
        subject_dirs = [d for d in os.listdir(self.data_dir)
                       if os.path.isdir(os.path.join(self.data_dir, d))
                       and (d.startswith('P') or d.startswith('M'))]
        subject_dirs = sorted(subject_dirs)

        if max_subjects:
            subject_dirs = subject_dirs[:max_subjects]

        print(f"\nÎncărcare date fMRI pentru {len(subject_dirs)} subiecți...")

        data = {exp: {'sentences': [], 'fmri': [], 'subjects': [], 'metadata': []}
                for exp in self.experiments.keys()}

        for subject in tqdm(subject_dirs, desc="Subiecți"):
            subject_path = os.path.join(self.data_dir, subject)

            for exp_name, exp_file in self.experiments.items():
                mat_path = os.path.join(subject_path, exp_file)

                if not os.path.exists(mat_path):
                    continue

                try:
                    mat_data = loadmat(mat_path)

                    fmri = mat_data['examples']

                    if len(data[exp_name]['sentences']) == 0:
                        sentences = mat_data['keySentences']
                        sentences = [s[0][0] if len(s[0]) > 0 else '' for s in sentences]
                        data[exp_name]['sentences'] = sentences

                    data[exp_name]['fmri'].append(fmri)
                    data[exp_name]['subjects'].append(subject)

                    if 'labelsPassageForEachSentence' in mat_data:
                        data[exp_name]['metadata'].append({
                            'passage_idx': mat_data['labelsPassageForEachSentence'][:, 0],
                            'passage_labels': mat_data.get('keyPassages', None)
                        })

                except Exception as e:
                    print(f"  Error loading {mat_path}: {e}")
                    continue

        processed_data = {}
        for exp_name, exp_data in data.items():
            if len(exp_data['fmri']) == 0:
                continue


            fmri_concat = np.concatenate(exp_data['fmri'], axis=1)

            voxel_to_subject = []
            for subj, fmri in zip(exp_data['subjects'], exp_data['fmri']):
                voxel_to_subject.extend([subj] * fmri.shape[1])

            processed_data[exp_name] = {
                'sentences': exp_data['sentences'],
                'fmri': fmri_concat,
                'subjects': exp_data['subjects'],
                'voxel_to_subject': np.array(voxel_to_subject),
                'n_voxels_per_subject': [f.shape[1] for f in exp_data['fmri']]
            }

            print(f"\n{exp_name}:")
            print(f"  Sentences: {len(exp_data['sentences'])}")
            print(f"  Subjects: {len(exp_data['subjects'])}")
            print(f"  fMRI shape: {fmri_concat.shape}")
            print(f"  Voxels per subject: {[f.shape[1] for f in exp_data['fmri']]}")

        return processed_data

    def get_combined_data(self, max_subjects=None):
        """
        Returnează date combinate pentru ambele experimente.

        Returns:
            sentences: Lista de propoziții (627 total)
            fmri: Dict {subject: {exp: fmri_array}}
            experiment_labels: Lista cu experimentul pentru fiecare propoziție
        """
        raw_data = self.load_fmri_data(max_subjects=max_subjects)

        all_sentences = []
        experiment_labels = []

        for exp_name in ['243sentences', '384sentences']:
            if exp_name in raw_data:
                all_sentences.extend(raw_data[exp_name]['sentences'])
                experiment_labels.extend([exp_name] * len(raw_data[exp_name]['sentences']))

        return {
            'sentences': all_sentences,
            'experiment_labels': experiment_labels,
            'raw_data': raw_data
        }
```

## 5. Metrica Encoding (Cross-Validated Regression)

```python
def cross_validated_encoding(X, Y, n_splits=5, alpha=1.0):
    """
    Cross-validated ridge regression cu Pearson correlation.
    Aceasta este metrica folosită în neural-nlp pentru Pereira2018-encoding.

    Args:
        X: Activări model [n_samples, n_features]
        Y: Date neurale [n_samples, n_voxels]
        n_splits: Număr de fold-uri CV
        alpha: Regularizare Ridge

    Returns:
        correlations: Array cu corelații per voxel
        mean_correlation: Media corelațiilor
    """
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    n_voxels = Y.shape[1]
    all_predictions = np.zeros_like(Y)

    for train_idx, test_idx in kf.split(X):
        X_train, X_test = X[train_idx], X[test_idx]
        Y_train, Y_test = Y[train_idx], Y[test_idx]

        # Fit Ridge regression
        model = Ridge(alpha=alpha)
        model.fit(X_train, Y_train)

        # Predict
        all_predictions[test_idx] = model.predict(X_test)

    # Calculează Pearson correlation per voxel
    correlations = []
    for v in range(n_voxels):
        if np.std(Y[:, v]) < 1e-10 or np.std(all_predictions[:, v]) < 1e-10:
            correlations.append(0)
        else:
            r, _ = pearsonr(Y[:, v], all_predictions[:, v])
            correlations.append(r if not np.isnan(r) else 0)

    correlations = np.array(correlations)
    return correlations, np.median(correlations)


def evaluate_model_per_subject(model_activations, pereira_data, experiment='243sentences'):
    """
    Evaluează un model per subiect (ca în neural-nlp).

    Returns:
        results: Dict cu scoruri per layer și per subiect
    """
    exp_data = pereira_data[experiment]
    fmri_list = []

    # Reconstruiește datele per subiect
    start_idx = 0
    subject_fmri = {}
    for subj, n_voxels in zip(exp_data['subjects'], exp_data['n_voxels_per_subject']):
        subject_fmri[subj] = exp_data['fmri'][:, start_idx:start_idx + n_voxels]
        start_idx += n_voxels

    results = {'per_layer': {}, 'per_subject': {}}

    for layer_idx, activations in tqdm(model_activations.items(), desc="Evaluating layers"):
        layer_scores = []
        subject_scores = {}

        for subj, fmri in subject_fmri.items():
            # Cross-validated encoding
            correlations, median_corr = cross_validated_encoding(activations, fmri)
            layer_scores.append(median_corr)
            subject_scores[subj] = {
                'median': median_corr,
                'mean': np.mean(correlations),
                'correlations': correlations
            }

        # Agregare: median across subjects (ca în neural-nlp)
        results['per_layer'][layer_idx] = {
            'median_across_subjects': np.median(layer_scores),
            'mean_across_subjects': np.mean(layer_scores),
            'std_across_subjects': np.std(layer_scores),
            'per_subject': subject_scores
        }

    return results


def evaluate_model_combined(model_activations, pereira_data, experiment='243sentences'):
    """
    Evaluează un model pe datele combinate (toți voxelii).
    Mai rapid, dar mai puțin precis decât per-subject.
    """
    exp_data = pereira_data[experiment]
    Y = exp_data['fmri']  # [n_sentences, n_voxels_total]

    results = {}

    for layer_idx, X in tqdm(model_activations.items(), desc="Evaluating layers"):
        correlations, median_corr = cross_validated_encoding(X, Y)

        results[layer_idx] = {
            'median': median_corr,
            'mean': np.mean(correlations),
            'std': np.std(correlations)
        }

    return results
```

## 7. Vizualizare Rezultate

```python
def plot_results(all_results, title="Model Comparison on Pereira2018"):
    """
    Plotează rezultatele.
    """
    if not all_results:
        print("No results to plot!")
        return

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Plot 1: Score per layer (normalized depth)
    ax1 = axes[0]
    for model_name, results in all_results.items():
        layers = sorted(results.keys())
        scores = [results[l]['median'] for l in layers]
        # Normalizează la 0-1 pentru comparație
        norm_layers = np.linspace(0, 1, len(layers))
        ax1.plot(norm_layers, scores, marker='o', label=model_name, linewidth=2, markersize=4)

    ax1.set_xlabel('Normalized Layer Depth (0=embedding, 1=final)', fontsize=12)
    ax1.set_ylabel('Median Pearson Correlation', fontsize=12)
    ax1.set_title('Encoding Score per Layer', fontsize=14)
    ax1.legend(loc='best')
    ax1.grid(True, alpha=0.3)
    ax1.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

    # Plot 2: Best score per model
    ax2 = axes[1]
    model_names = list(all_results.keys())
    best_scores = []
    best_layers = []

    for model_name in model_names:
        results = all_results[model_name]
        best_layer = max(results.keys(), key=lambda l: results[l]['median'])
        best_scores.append(results[best_layer]['median'])
        best_layers.append(best_layer)

    colors = plt.cm.Set2(np.linspace(0, 1, len(model_names)))
    bars = ax2.bar(model_names, best_scores, color=colors, edgecolor='black')

    # Adaugă valorile pe bare
    for bar, score, layer in zip(bars, best_scores, best_layers):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2, height + 0.005,
                f'{score:.3f}\n(L{layer})', ha='center', va='bottom', fontsize=9)

    ax2.set_ylabel('Best Median Pearson Correlation', fontsize=12)
    ax2.set_title('Best Encoding Score per Model', fontsize=14)
    ax2.tick_params(axis='x', rotation=45)
    ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

    plt.suptitle(f"{title} ({EXPERIMENT})", fontsize=16, y=1.02)
    plt.tight_layout()
    plt.savefig('pereira2018_model_comparison.png', dpi=150, bbox_inches='tight')
    plt.show()

    return fig

if all_results:
    plot_results(all_results)
```

```python
def create_results_table(all_results):
    """
    Creează un tabel cu rezultatele.
    """
    if not all_results:
        print("No results!")
        return None

    rows = []

    for model_name, results in all_results.items():
        best_layer = max(results.keys(), key=lambda l: results[l]['median'])
        best_score = results[best_layer]['median']

        # Scor la layer-ul middle
        middle_layer = len(results) // 2
        middle_score = results[middle_layer]['median']

        # Media pe toate layerele
        all_medians = [results[l]['median'] for l in results.keys()]
        mean_score = np.mean(all_medians)

        rows.append({
            'Model': model_name,
            'Best Layer': best_layer,
            'Best Score': f"{best_score:.4f}",
            'Middle Layer Score': f"{middle_score:.4f}",
            'Mean Score': f"{mean_score:.4f}",
            'Num Layers': len(results)
        })

    df = pd.DataFrame(rows)
    df = df.sort_values('Best Score', ascending=False)
    return df

if all_results:
    results_df = create_results_table(all_results)
    print("\n" + "="*70)
    print(f"REZULTATE FINALE - Pereira2018 Encoding ({EXPERIMENT})")
    print("="*70)
    print(results_df.to_string(index=False))
```

## 8. Salvare Rezultate

```python
# Salvează rezultatele
def save_results(all_results, filename='pereira2018_results'):
    # JSON (fără numpy arrays)
    json_results = {}
    for model, results in all_results.items():
        json_results[model] = {}
        for layer, data in results.items():
            json_results[model][str(layer)] = {
                'median': float(data['median']),
                'mean': float(data['mean']),
                'std': float(data['std'])
            }

    with open(f'{filename}.json', 'w') as f:
        json.dump(json_results, f, indent=2)

    # Pickle (complet)
    with open(f'{filename}.pkl', 'wb') as f:
        pickle.dump(all_results, f)

    print(f"Results saved to {filename}.json and {filename}.pkl")

if all_results:
    save_results(all_results, f'pereira2018_{EXPERIMENT}_results')
```
